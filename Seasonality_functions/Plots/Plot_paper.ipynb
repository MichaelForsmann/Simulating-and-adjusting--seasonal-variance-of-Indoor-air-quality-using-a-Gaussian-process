```python
import torch
import arviz as az
from scipy.optimize import curve_fit
from bokeh.plotting import figure, show,output_file, save
from bokeh.transform import factor_cmap, factor_mark
from bokeh.palettes import Spectral
from bokeh.models import Slope, Div
from bokeh.io import curdoc,output_notebook,export_png
from bokeh.layouts import column,gridplot
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from bokeh.models import Band, ColumnDataSource
import pandas as pd
from Plotting import *
```

```python
data=pd.read_excel(r"../../Data_and_Preprocessing/before_season_correction.xlsx")

train=data.loc[data.label==0,:]
test=data.loc[data.label==1,:]
```

```python
data_pm=pd.read_excel(r"../../Data_and_Preprocessing/pm25_BC_corrected.xlsx")
train_pm=data_pm.loc[data_pm.label==0,:]
test_pm=data_pm.loc[data_pm.label==1,:]

device="cuda"
```

```python
r2_score(data_pm.pm25_Gaussion.values,gpr_no2(torch.tensor(data_pm.corrected_week.values).float())[0].cpu.detach().numpy())
```

```python
gpr_no2=torch.load("../models/no2_seasonality",weights_only=False,map_location='cuda')
gpr_nox=torch.load("../models/nox_seasonality",weights_only=False,map_location='cuda')
gpr=torch.load("../models/pm25_seasonality",weights_only=False,map_location='cuda')
gpr_bc=torch.load("../models/bc_seasonality",weights_only=False,map_location='cuda')
gpr_acetone=torch.load("../models/acetone_seasonality",weights_only=False,map_location='cuda')
gpr_form=torch.load("../models/formaldehyd_seasonality",weights_only=False,map_location='cuda')
gpr_acetald=torch.load("../models/acetald_seasonality",weights_only=False,map_location='cuda')
gpr_vocs=torch.load("../models/vocs_seasonality",weights_only=False,map_location='cuda')
```

```python
output_notebook()
p=plot("pm25_Gaussion",r"$$PM_{2.5}$$ Seasonality",train_pm,test_pm,gpr,x_label=None)
p1=plot("BC_Gaussion",r"Balck Carbon Seasonality",train_pm,test_pm,gpr_bc,diffrence=0.3,x_label=None,y_label=None)
p2=plot("no2",r"$$NO_{2}$$ Seasonality",train,test,gpr_no2,x_label=None,diffrence=8)
p3=plot("nox",r"$$NO_{x}$$ Seasonality",train,test,gpr_nox,x_label=None,y_label=None)
p4=plot("formaldehyd",r"Formaldehyd Seasonality",train,test,gpr_form,x_label=None)
p5=plot("acetald",r"Acetald Seasonality",train,test,gpr_acetald,x_label=None,y_label=None)
p6=plot("acetone",r"Acetone Seasonality",train,test,gpr_acetone)
p7=plot("Total_VOCs",r"VOCs Seasonality",train,test,gpr_vocs,y_label=None)

grid =  gridplot([[p, p1],[p2,p3],[p4,p5],[p6,p7]], width=1000, height=500);
show(column(grid));
```
