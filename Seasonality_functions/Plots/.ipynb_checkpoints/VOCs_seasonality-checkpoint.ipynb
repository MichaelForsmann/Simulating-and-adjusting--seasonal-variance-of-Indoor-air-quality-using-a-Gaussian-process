```python
import torch
import arviz as az
from scipy.optimize import curve_fit
from bokeh.plotting import figure, show,output_file, save
from bokeh.transform import factor_cmap, factor_mark
from bokeh.palettes import Spectral
from bokeh.models import Slope, Div
from bokeh.io import curdoc,output_notebook,export_png
from bokeh.layouts import column,gridplot
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from bokeh.models import Band, ColumnDataSource
import pandas as pd
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
from Plotting import *
```

```python
data=pd.read_excel(r"../../Data_and_Preprocessing/before_season_correction.xlsx")
```

```python
train=data.loc[data.label==0,:]
test=data.loc[data.label==1,:]
```

```python
train
```

```python
gpr_acetone=torch.load("../models/acetone_seasonality",weights_only=False,map_location='cuda')
gpr_form=torch.load("../models/formaldehyd_seasonality",weights_only=False,map_location='cuda')
gpr_acetald=torch.load("../models/acetald_seasonality",weights_only=False,map_location='cuda')
gpr_vocs=torch.load("../models/vocs_seasonality",weights_only=False,map_location='cuda')
```

```python
mean_form_train,data_form_train=roll_week(train,"formaldehyd")
mean_acetald_train,data_acetald_train=roll_week(train,"acetald")
mean_acetone_train,data_acetone_train=roll_week(train,"acetone")
mean_total_vocs_train,data_total_vocs_train=roll_week(train,"Total_VOCs")
```

```python
mean_form_test,data_form_test=roll_week(test,"formaldehyd")
mean_acetald_test,data_acetald_test=roll_week(test,"acetald")
mean_acetone_test,data_acetone_test=roll_week(test,"acetone")
mean_total_vocs_test,data_total_vocs_test=roll_week(test,"Total_VOCs")
```

```python
plot_form,model_form=gaussian_plot(train,gpr_form) 
plot_acetald,model_acetald=gaussian_plot(train,gpr_acetald) 
plot_acetone,model_acetone=gaussian_plot(train,gpr_acetone) 
plot_vocs,model_vocs=gaussian_plot(train,gpr_vocs) 
```

```python
plot_form
```

```python
r2_form,r2_acetald,r2_acetone=r2_score(data_form_train.y,model_form.y),r2_score(data_acetald_train.y,model_acetald.y),r2_score(data_acetone_train.y,model_acetone.y)
r2_form_test,r2_acetald_test,r2_acetone_test=r2_score(data_form_test.y,model_form.y),r2_score(data_acetald_test.y,model_acetald.y),r2_score(data_acetone_test.y,model_acetone.y)
r2_vocs,r2_vocs_test=r2_score(data_vocs_train.y,model_vocs.y),r2_score(data_total_vocs_test.y,model_vocs.y)
```

```python
r2_form_train_vs_test=r2_score(data_form_train.y,data_form_test.y)
r2_acetald_train_vs_test=r2_score(data_acetald_train.y,data_acetald_test.y)
r2_acetone_train_vs_test=r2_score(data_acetone_train.y,data_acetone_test.y)
r2_vocs_train_vs_test=r2_score(data_vocs_train.y,data_total_vocs_test.y)

r2_form_train_vs_test,r2_acetald_train_vs_test,r2_acetone_train_vs_test,r2_vocs_train_vs_test
```

```python
def plot(particle,train,test,model,x_label=r"$$Week \ of \ the \ year$$",y_label=r"$$\frac{\mu g}{m^3} $$"):
    TOOLS="hover,crosshair,pan,wheel_zoom,zoom_in,zoom_out,box_zoom,undo,redo,reset,tap,save,box_select,poly_select,lasso_select,examine,help"
    mean_train,roll_train=roll_week(train,particle)
    mean_test,roll_test=roll_week(test,particle)
    model_plot,model_dataframe=gaussian_plot(train,model) 
    r2_train,r2_test=r2_score(roll_train.y,model_dataframe.y),r2_score(roll_test.y,model_dataframe.y)
    p = figure(tools=TOOLS,x_range=(model_dataframe.index.min(), model_dataframe.index.max()),y_range=(roll_train.y.min()-5,roll_train.y.max()+5));
    p.title.text_font_size = '13pt'
    p.title.text = particle+" Seasonality";  
    p.ygrid.grid_line_alpha=0.5;
    p.line(roll_train.index, roll_train.y, line_width=3,color="green",legend_label="Train roll mean R2 "+str(round(r2_train,2)));
    p.line(roll_test.index, roll_test.y, line_width=3,color="orange",legend_label="Test roll mean R2 "+str(round(r2_test,2)));
    p.line(model_dataframe.index, model_dataframe.y, line_width=3,color="red",legend_label="Gaussion model ");
    p.scatter(train.corrected_week, y=train[particle], color="blue", marker="dot", size=20, alpha=0.4,legend_label="raw points");
    band_data= Band(base="index", lower="lower", upper="upper",source=mean_train, fill_color="red", line_color="black",fill_alpha=0.2);
    band_model = Band(base="index", lower="lower", upper="upper",source=model_plot,fill_alpha=0.5, fill_color="blue", line_color="black");
    p.yaxis.axis_label_orientation  = 0
    p.add_layout(band_data);
    p.add_layout(band_model);
    p.xaxis.axis_label = x_label
    p.yaxis.axis_label = y_label
    p.xaxis.axis_label_text_font_size = "18px";
    p.yaxis.axis_label_text_font_size = "18px";
    p.legend.label_text_font_size = '13pt'
    return p
```

```python
p=plot("formaldehyd",train,test,gpr_form)
p1=plot("acetald",train,test,gpr_acetald)
p2=plot("acetone",train,test,gpr_acetone)
p3=plot("Total_VOCs",train,test,gpr_vocs)

grid =  gridplot([[p, p1],[p2,p3]], width=500, height=500);
show(column(grid));
```
