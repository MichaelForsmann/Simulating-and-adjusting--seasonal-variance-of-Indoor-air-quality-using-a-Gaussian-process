```python
import pandas as pd
import arviz as az
from Functions import *
```

```python
df = pd.read_excel('Adjusting/IAQ_third_data.xlsx', index_col=0)
```

```python
allergies=df.loc[:,['fel_d_1_dust', 'der_f_1_dust',
       'der_p_1_dust', 'can_f_1_dust']]
```

```python
data=pd.read_excel(r"../Data_and_Preprocessing/before_season_correction.xlsx")
device='cuda'
train=data.loc[data.label==0,:,]
```

```python
week=data.loc[data.maalnr==1,:].set_index("cpnr").corrected_week
```

```python
allergines_ll=pd.concat([allergies,week],axis=1)
```

```python
allergines_ll=allergines_ll.dropna(subset="corrected_week")
```

```python
allergines_ll=make_periodic(allergines_ll,"corrected_week",device) 
```

```python
allergines_ll.loc[:,["fel_d_1_dust","der_f_1_dust","der_p_1_dust","can_f_1_dust"]]=allergines_ll.loc[:,["fel_d_1_dust","der_f_1_dust","der_p_1_dust","can_f_1_dust"]]/allergines_ll.loc[:,["fel_d_1_dust","der_f_1_dust","der_p_1_dust","can_f_1_dust"]].mean()
```

```python
X_flow,y_flow=make_Xy(allergines_ll,["cos_theta","sin_theta"],"der_f_1_dust",device) 
nuts_kernel_flow,gpr_flow=model(X_flow,y_flow,device) 
pyro_data_flow,gpr_flow=train_model(X_flow,nuts_kernel_flow,gpr_flow,device) 
torch.save(gpr_flow,"models/der_f_1_dust");
```

```python

X_flow,y_flow=make_Xy(allergines_ll,["cos_theta","sin_theta"],"der_p_1_dust",device) 
nuts_kernel_flow,gpr_flow=model(X_flow,y_flow,device) 
pyro_data_flow,gpr_flow=train_model(X_flow,nuts_kernel_flow,gpr_flow,device) 
torch.save(gpr_flow,"models/allergy_seasonality");
```

```python
allergines_ll.to_excel("allergines_ll.xlsx")
```
