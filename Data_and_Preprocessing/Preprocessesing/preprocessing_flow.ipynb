```python
from sklearn.experimental import enable_iterative_imputer
import pandas as pd
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import StandardScaler
import numpy as np
import kennard_stone as ks
```

```python
PM25=pd.read_excel("../pm_flow.xlsx")
```

```python
scaler = StandardScaler()
imputer = IterativeImputer(random_state=0)
pm25=imputer.fit_transform(scaler.fit_transform(PM25))
pm25=scaler.inverse_transform(pm25)
PM25=pd.DataFrame(np.round(pm25,3),columns=PM25.columns,index=PM25.index)
```

```python
bnr=PM25.reset_index().loc[:,["household","Area","passive","flow","road"]].groupby("household").mean().reset_index()
X_train, X_test, y_train, y_test = ks.train_test_split(bnr.loc[:,["Area","passive","flow","road"]], bnr.household, test_size = 0.3);
PM25=PM25.set_index("household")
train=PM25.loc[y_train.values,:]
test=PM25.loc[y_test.values,:]
```

```python
PM25.loc[train.index,"label"]=0
PM25.loc[test.index,"label"]=1
```

```python
PM25.to_excel("../Pre_flow_cal.xlsx")
```
