```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import torch
import kennard_stone as ks
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
no2=pd.read_excel(r"../Chemicals_corrected.xlsx")
data=pd.read_excel(r"../Pre_flow_cal.xlsx")
```

```python
# Extract week,dat
no2["sadto"]=pd.to_datetime(no2.start_n)
no2=no2.assign(week=no2.sadto.apply(lambda x: x.week))
no2=no2.assign(month=no2.sadto.apply(lambda x: x.month))
no2=no2.assign(year=no2.sadto.apply(lambda x: x.year))
no2=no2.assign(day=no2.sadto.apply(lambda x: x.day_of_year))
```

```python
data_bnr=data.set_index("bnr")
data_bnr=data_bnr.drop(['flow', 'pm_final', 'sv_final','week', 'month', 'year', 'day',
       'corrected_week', 'label', 'BC_Gaussion', 'BC_linear', 'BC_exp',
       'BC_log', 'pm25_Gaussion', 'pm25_linear', 'pm25_exp', 'pm25_log'],axis=1)
data_bnr.index.name="cpnr"
data_bnr_1=data_bnr.loc[data_bnr.maalnr==1,:]
data_bnr_2=data_bnr.loc[data_bnr.maalnr==2,:]
data_bnr_3=data_bnr.loc[data_bnr.maalnr==3,:]
```

```python
#removeig unwanted stuff
no2=no2.replace(",",".",regex=True).rename(columns={"no2_(μg/m3)":"no2","no_(μg/m3)":"Nox","formald_(μg/m3)":"formaldehyd","acetald_(μg/m3)":"acetald","acetone_(μg/m3)":"acetone"})
no2.no2=no2.no2.astype(float)
no2.Nox=no2.Nox.astype(float)
```

```python
no2.loc[no2.visit=="6mdr","visit"]=1
no2.loc[no2.visit=="1yr","visit"]=2
no2.loc[no2.visit=="18mdr","visit"]=3
no2.loc[no2.visit=="2yr","visit"]=4
no2.loc[no2.visit=="30mdr","visit"]=5
no2.loc[no2.visit==18,"visit"]=5
no2.loc[no2.visit=="?","visit"]=4
```

```python
no2["corrected_week"]=no2.day/7
no2=no2.drop(['sadto', 'week', 'month','batch_c', 'start_c', 'stop_c',
       'batch_n', 'start_n', 'stop_n','Unnamed: 0','year','day','Nox'],axis=1)
```

```python
no2_cpnr=no2.set_index("cpnr")
no2_cpnr_1=no2_cpnr.loc[no2_cpnr.visit==1]
no2_cpnr_2=no2_cpnr.loc[no2_cpnr.visit==2]
no2_cpnr_3=no2_cpnr.loc[no2_cpnr.visit==3]
no2_cpnr_4=no2_cpnr.loc[no2_cpnr.visit==4]
no2_cpnr_5=no2_cpnr.loc[no2_cpnr.visit==5]
visit=pd.concat([data_bnr_1,no2_cpnr_1],axis=1).dropna(subset="no2").reset_index()
visit2=pd.concat([data_bnr_2,no2_cpnr_2],axis=1).dropna(subset="no2").reset_index()
visit3=pd.concat([data_bnr_3,no2_cpnr_3],axis=1).dropna(subset="no2").reset_index()
visit4=pd.concat([data_bnr_3,no2_cpnr_4],axis=1).dropna(subset="no2").reset_index()
visit5=pd.concat([data_bnr_3,no2_cpnr_5],axis=1).dropna(subset="no2").reset_index()

full_data_no2=pd.concat([visit,visit2,visit3,visit4,visit5])
```

```python
full_data_no2=full_data_no2.drop(["no2_corrected","formaldehyd_corrected"],axis=1)
```

```python
labels=np.array(['constrution_year', 'Area', 'floor_level',
       'powerplant', 'road', 'motorway', 'forest', 'coastline', 'airport',
       'farmland', 'industrial', 'carpet', 'cookerhood', 'passive', 'gastove',
       'fireplace'])
```

```python
impute=full_data_no2.loc[:,labels].copy()
```

```python
imputer = IterativeImputer(random_state=0,skip_complete=True,max_iter=50)
no2_1=imputer.fit_transform(scaler.fit_transform(impute))
no2_2=scaler.inverse_transform(no2_1)
no2_df=pd.DataFrame(np.round(no2_2,1),columns=impute.columns,index=full_data_no2.cpnr)
```

```python
full_data_no2.loc[:,labels]=no2_df.values
```

```python
cpnr=full_data_no2.loc[:,["corrected_week","gastove","passive","road","cpnr"]].groupby("cpnr").mean().reset_index()
X_train, X_test, y_train, y_test = ks.train_test_split(cpnr.loc[:,["gastove","passive","corrected_week","road"]], cpnr.cpnr, test_size = 0.4);
full_data_no2_1=full_data_no2.set_index("cpnr")
train=full_data_no2_1.loc[y_train.values,:].reset_index()
test=full_data_no2_1.loc[y_test.values,:].reset_index()
```

```python
train["label"]=0
test["label"]=1
full=pd.concat([train,test])
```

```python
full["Total_VOCs"]=full.loc[:,["formaldehyd","acetald","acetone"]].sum(axis=1)
```

```python
full.to_excel("../before_season_correction.xlsx")
```
